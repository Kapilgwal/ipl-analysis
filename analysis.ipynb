```python
def team_renaming(text):
    
    if text == 'Delhi Daredevils':
        text = 'Delhi Capitals'
        
    if text == 'Rising Pune Supergiants':
        text = 'Rising Pune Supergiant'
        
    if text == 'Kings XI Punjab':
        text = 'Punjab Kings'     
        
    if text == 'Royal Challengers Bangalore':
        text = 'Royal Challengers Bengaluru'     
    
    return text       
```

```python
df['team1'] = df['team1'].apply(team_renaming)
df['team2'] = df['team2'].apply(team_renaming)
df['toss_winner'] = df['toss_winner'].apply(team_renaming)
df['winner'] = df['winner'].apply(team_renaming)
```

```python

def match_type_renaming(text):
    
    if text == 'Semi Final':
        text = 'Eliminator'
        
    if text == '3rd Place Play-Off':   
        text = 'Eliminator'
    
    if text == 'Elimination Final':
        text = 'Qualifier 2'
        
    return text       
```

```python
df['match_type'] = df['match_type'].apply(match_type_renaming)
```

```python
df['match_type'].value_counts()
```

```python
def city_renaming(text):
    
    if text == 'Bangalore':
        text = 'Bengaluru'
    
    if text == 'Navi Mumbai':
        text = 'Mumbai'    
    
    return text       
```

```python
df['city'] = df['city'].apply(city_renaming)
```

```python
mask = df['winner'] == 'Mumbai Indians'
team_df = df[mask]
win_grounds = team_df.groupby('season')
win_grounds['player_of_match'].value_counts().sort_values(ascending=False)
```

```python
def top_underperformed(team):
    mask1 = df['team1'] == team
    mask2 = df['team2'] == team
    team_df = df[mask1 | mask2]
    defeated = team_df['winner']
    win_grounds = team_df.groupby('season')
    print(win_grounds.value_counts().sort_values(ascending=False))
```
